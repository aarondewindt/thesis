```python
import gym
import ray
import numpy as np
import ipywidgets as widgets
from ray.rllib.agents import ppo
from tqdm.auto import trange

from cw.vdom import hyr
```

```python
ray.init()
```

```python
trainer = ppo.PPOTrainer(env="CartPole-v1")
```

```python
training_history = {
    "episode_reward_max": [],
    "episode_reward_min": [],
    "episode_reward_mean": [],
    "episode_len_mean": []
}
```

```python
out = widgets.Output()
display(out)
with out:
    display("")

for i in trange(10):
    r = trainer.train()
    
    out.clear_output(True)
    with out:
        display(hyr({k: r[k] for k in training_history.keys()}))
        
    for k in training_history.keys():
        training_history[k].append(r[k])
    
```

```python
env = gym.make("CartPole-v1")

# run until episode ends
episode_reward = 0
done = False
obs = env.reset()
while not done:
    action = trainer.compute_action(obs)
    obs, reward, done, info = env.step(action)
    episode_reward += reward
    env.render()

print(episode_reward)
```

```python
env.close()
```
