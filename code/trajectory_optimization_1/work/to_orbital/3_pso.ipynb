```python
def generate_result_path(results_path: Path, note: str):
    name = datetime.now().strftime(f'%Y_%m_%d_%H_%M_%S_%f__{note}')
    path = (results_path / name).resolve()
    path.mkdir()
    return path


results_path = Path('./results') / "pso"
results_path.mkdir(exist_ok=True)
```

```python
case_runner = CaseRunner(
    env=env,
    max_time=1000,
    n_checkpoints=4,
    results_path=generate_result_path(results_path, "opt_2")
)
```

```python
optimizer = PSOOptimizer(
    case_runner=case_runner,
    n_particles=100
)
```

```python
optimizer.optimize(
    iters=1000,
    n_processes=8
)
```
